```python
angle = 74
data = ndimage.rotate(im_png, angle=angle)
plt.imshow(data)
```

```python
sum_array = data.sum(axis=2, keepdims=True)
sum_array = sum_array / np.amax(sum_array)

sum_array.shape
```
